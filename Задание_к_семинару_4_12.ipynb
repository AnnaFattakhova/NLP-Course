**Задание**

Представьте, что вы работаете аналитиком в проекте по разработке системы машинного перевода. Ваша команда создает систему машинного перевода на основе правил. Прежде чем ваши коллеги приступят к разработке системы, вам, как аналитику, нужно описать архитектуру будущей системы.

Опишите структуру своей системы машинного перевода на основе правила в формате псевдокода, алгоритма или блок-схемы (на выбор). В процессе работы вам нужно будет ответить на следующие вопросы:


*   *Какую языковую пару обрабатывает ваша система?*
*   *Какие виды анализа и предварительной обработки данных вы применяете?*
*   *Какие языковые правила вам нужно описать?*
*   *Как ваша система будет генерировать результат, т.е. перевод?*

Дополнительно попробуйте ответить на вопрос:
*   *Какие библиотеки NLP вы будете использовать?*









**Решение**

I. Языковая пара
Русский → Испанский.

II.
1) Принимаем на вход текст пользователя
2) Токенизируем (гибридная: по словам, для неизвестных слов посимвольно)
3) Лемматизируем (при помощи SpaCy)
4) Проводим анализ текста на исходном языке: лексический (значения слов), морфологический (грамматические характеристики каждого исходного слова), синтаксический (синтаксическая структура предложения: возможно построение дерева зависимостей)
4) Применяем языковые правила
5) Генерируем первод на целевом языке с учетом всех правил
6) Вывод текста


III. Языковые правила для перевода:

- транслитерация для имен собственных, узкоспецифических терминов и дргуих незнакомых слов (например, с опечатками)
- перевод устойчивых выражений (идиом)
- лексика (слова должны заменяться на соответствующие в целевом языке)
- морфология (спряжение глаголов - времена, лица, числа; артикли и их согласование с родом и числом существительных "S(m, SG) → 'el' + S"; согласование прилагательных с существительными ...)
- синтаксис (порядок слов "SVO → SOV", опускание личных местоимений при глаголах "P(1, SG)V  → V" ...)
- пунктуация (в испанском требуются дополнительные знаки в началах вопросительных и восклицательных предожений: ¿, ¡)
...

IV. Генерация текста: леммы исходного языка сопоставляются с леммами целевого, затем применябтся заданные языковые правила, добавляется пунктуация. Затем происходит постобработка текста.
