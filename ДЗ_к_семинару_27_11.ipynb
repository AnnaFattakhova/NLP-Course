```python
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec
import re
```

```python
with open('датасет.txt', 'r', encoding='utf-8') as f:
    text = f.read()
    text = re.sub(r'[^\w\s\.\?!]', '', text)
```

```python
import spacy
```

```python
load_model = spacy.load('en_core_web_sm', disable = ['parser','ner'])
```

```python
doc = load_model(text)
```

```python
new_text = " ".join([token.lemma_ for token in doc])
```

```python
sentences = [simple_preprocess(sentence) for sentence in new_text.splitlines() if sentence]
```

```python
sentences
```

```python
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
sentences = [
    [word for word in simple_preprocess(sentence) if word not in stop_words]
    for sentence in new_text.splitlines() if sentence
]
```

```python
model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)
```

```python
model.wv.most_similar('dragon', topn=20)
```

```python
from gensim.corpora.dictionary import Dictionary
from gensim.models import LdaModel
```

```python
dictionary = Dictionary(sentences)
corpus = [dictionary.doc2bow(sentence) for sentence in sentences]
```

```python
lda_model = LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15)
```

```python
topics = lda_model.print_topics(num_words=5)
for idx, topic in topics:
    print(f"Topic {idx}: {topic}")
```
