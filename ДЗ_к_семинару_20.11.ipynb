```python
import pandas as pd
import seaborn as sns #для визуализации
import matplotlib.pyplot as plt
from collections import defaultdict #для матрицы, популяции, статистики
import re
```

```python
file_path = 'датасет.txt'
```

```python
with open(file_path, 'r', encoding='utf-8') as file:
  text = file.read()
```

```python
text = text.lower()
text[500:1000]
```

```python
text = re.sub(r'[^\w\s\.\?!]', '', text)
text[500:1000]
```

```python
sentences = re.split(r'[\.\?!]', text) #делим на предложение по точкам, вопросам, восклицательным знакам
sentences[1:3]
```

```python
sentences = [sentence.split() for sentence in sentences if sentence]
sentences[3]
```

```python
co_occurrence = defaultdict(lambda: defaultdict(int))
```

```python
#чтобы сделать матрицу менее разрешенной, можно увеличить до 10 размер окна


def update_co_occurrence(sentence, window_size=1):
    words = ["<START>"] + sentence + ["<END>"]
    for i, word in enumerate(words):
        for j in range(max(0, i - window_size), min(len(words), i + window_size + 10)):
            if i != j:
                co_occurrence[word][words[j]] += 1

for sentence in sentences:
    update_co_occurrence(sentence)

co_occurrence['fire']
```

```python
tokens = list(co_occurrence.keys())
tokens[:10]
```

```python
co_occurrence_df = pd.DataFrame(co_occurrence).fillna(0).astype(int).reindex(tokens, columns=tokens, fill_value=0)
co_occurrence_df
```
